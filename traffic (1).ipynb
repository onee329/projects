```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
traffic = pd.read_csv('Traffic-Prediction-using-Machine-Learning.csv')
traffic
```

```python
traffic.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
traffic[['Date', 'Day']] = encoder.fit_transform(traffic[['Date', 'Day']])
```

```python
traffic
```

```python
traffic['CodedDay'] = traffic['CodedDay'].astype(int)
traffic['day_sin'] = np.sin(2 * np.pi * traffic['CodedDay'] / 7)
traffic['day_cos'] = np.cos(2 * np.pi * traffic['CodedDay'] / 7)
```

```python
traffic.info()
```

```python
x = traffic.iloc[:,2:6].values
y = traffic.iloc[:,6:7].values
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
Sc_x = StandardScaler()
xtrain  = Sc_x.fit_transform(xtrain)
xtest = Sc_x.fit_transform(xtest)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 300, random_state = 0)
regressor.fit(xtrain,ytrain)
```

```python

y_pred = regressor.predict(xtest)
```

```python
if(y_pred.all()<2.5):
    y_pred=np.round(y_pred-0.5)
else:
    y_pred=np.round(y_pred+0.5)
```

```python
df1=(y_pred-ytest)/ytest
df1=round(df1.mean()*100,2)
print("Error = ", df1, "%")
a=100-df1
print("Accuracy =", a, "%")
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(xtrain,ytrain)
```

```python
y_pred = regressor.predict(xtest)
y_pred
```

```python
if(y_pred.all()<2.5):
    y_pred=np.round(y_pred-0.5)
    
else:
    y_pred=np.round(y_pred+0.5)

y_pred
```
